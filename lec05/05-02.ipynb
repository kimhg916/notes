# 5.2: Geometry of lines and planes

Let us now use the tools we have built so far to explore the geometry of affine lines and planes in $\mathbb{R}^3$. We will give several examples of different problems involving affine lines and planes.

### Lines through two points

Geometrically, two different points in $\mathbb{R}^3$ determine a line. Suppose we are given two points
$$ A(3,0,1),\ B(2,-1,1).$$
How do we find the line that goes through these two points?

Let us express the points with position vectors $\vec a$ and $\vec b$. The line passes through $\vec a$ and is pointing in the direction of
$$ \vec v = \vec b - \vec a = \begin{bmatrix} 2 \\ -1 \\ 1 \end{bmatrix} - \begin{bmatrix} 3 \\ 0 \\ 1 \end{bmatrix} = \begin{bmatrix} -1 \\ -1 \\ 0 \end{bmatrix}.$$
Hence we can express the line in terms of $\vec r = \begin{bmatrix} x \\ y \\ z \end{bmatrix}$ in the vector-parametric form
$$ \vec r = \vec a + t(\vec b - \vec a) = \begin{bmatrix} 3 \\ 0 \\ 1 \end{bmatrix} + t\begin{bmatrix} -1 \\ -1 \\ 0 \end{bmatrix} = \begin{bmatrix} 3 - t \\ -t \\ 1 \end{bmatrix}.$$
Expanding out each entry, we obtain the scalar-parametric form
$$ \begin{cases}x = 3 - t \\ y = -t \\ z = 1 \end{cases}$$
The symmetric form is obtained by solving for the parameter $t$:
$$ \frac{x-3}{-1} = \frac{y-0}{-1}, \ z = 1.$$

### Parallel lines

Suppose we are given the line $l$ in the following scalar-parametric form:
$$ l:\ x = 1 + t,\ y = 2,\ z = 3+2t.$$
How can we find a line $l'$ that is parallel to $l$ and passes through the point $P(2,4,1)$?

Note that we can read off the directional vector of the line $l$ from the scalar-parametric form as $\vec{v} = \begin{bmatrix} 1 \\ 0 \\ 2 \end{bmatrix}$. More precisely, $l$ is pointing in the direction of $\vec{v}$ and passes through the point $(1,2,3)$. Since $l'$ is parallel to $l$, the two lines will have the same directional vector. Thus the line $l'$ can be expressed as:
$$ l':\ x = 2 + t,\ y = 4,\ z = 1 + 2t. $$

### Planes through three points

Similarly to the case of a line through two points, we can find the plane that goes through three points in $\mathbb{R}^3$. Suppose we are given the following three points in $\mathbb{R}^3$:
$$ A(-1,2,1),\ B(2,-1,0),\ C(3,0,0).$$
How do we find the plane that goes through all three of these points?

Let us express these points with the position vectors $\vec a$, $\vec b$, and $\vec c$. Note that the plane passes through $\vec a$ and is spanned by the two vectors
$$\vec b - \vec a = \begin{bmatrix} 3 \\ -3 \\ -1 \end{bmatrix},\quad \vec c - \vec a = \begin{bmatrix} 4 \\ -2 \\ -1\end{bmatrix}.$$
Hence we can express the plane in scalar-parametric form as
$$ \begin{cases} x = -1 + 3t + 4s\\ y = 2 -3t -2s \\ z = -1 - t - s \end{cases}$$
Suppose we want to convert this into point-normal form. We find the normal vector using the cross product:
$$ \vec n = (\vec b - \vec a) \times (\vec c - \vec a) = \begin{bmatrix} 3 \\ -3 \\ -1 \end{bmatrix} \times \begin{bmatrix} 4 \\ -2 \\ -1\end{bmatrix} = \begin{bmatrix} 1 \\ -1 \\ 6 \end{bmatrix}.$$
Thus the point-normal form is given by
$$ \vec n\cdot (\vec r - \vec a) = \begin{bmatrix} 1 \\ -1 \\ 6 \end{bmatrix} \cdot \left(\begin{bmatrix} x \\ y \\ z \end{bmatrix} - \begin{bmatrix} -1 \\ 2 \\ 1 \end{bmatrix} \right) = 0.$$
Expanding out the dot product, we obtain the Cartesian form of the plane as
$$ (x+1) - (y-2) + 6(z-1) = 0,\quad x - y + 6z = 3.$$
We can verify that this plane indeed passes through the three points above by plugging in their coordinates to the Cartesian form above.

### Intersection of planes

Suppose we are given two planes in $\mathbb{R}^3$ through the origin:
$$ \begin{cases} P_1:\ x + y + z = 0 \\ P_2:\ -x + 2y = 0 \end{cases} $$
How do we find the intersection of these two planes?

We can read off the normal vectors of $P_1$, $P_2$ as 
$$ \vec n_1 = \begin{bmatrix} 1 \\ 1 \\ 1 \end{bmatrix},\quad \vec n_2 = \begin{bmatrix} -1 \\ 2 \\ 0 \end{bmatrix},$$
respectively. Geometrically, note that the intersection of the two planes should be a line through the origin whose directional vector $\vec v$ is perpendicular to both $\vec n_1$ and $\vec n_2$.

```python
import plotly.io as pio
import plotly.graph_objects as go
import numpy as np

# Set default renderer to high-DPI static image
pio.renderers.default = "png"

# ---
from utils import plot_vectors_non_origin

# 1. DEFINE VECTORS
O = np.array([0, 0, 0])

# Normal vector for Plane 1 (x + y + z = 0)
n1 = np.array([1.0, 1.0, 1.0])

# Normal vector for Plane 2 (-x + 2y = 0)
n2 = np.array([-1.0, 2.0, 0.0])

# Direction vector of the intersection line (v = n1 x n2)
v = np.cross(n1, n2)  # Results in [-2, -1, 3]

# Updated color assignments
color_p1 = 'blue'
color_p2 = 'green'
color_v = 'red'

vectors = [
    ((tuple(O), tuple(n1)), color_p1, ''),                 
    ((tuple(O), tuple(n2)), color_p2, ''),                
    ((tuple(O), tuple(v)), color_v, '')                    
]

# Generate the base lines
fig = plot_vectors_non_origin(vectors, vdeltax=0, vdeltay=0, vdeltaz=0)

# 2. ADD THE TWO PLANES
# Plane 1 (P1): z = -x - y
fig.add_trace(go.Mesh3d(
    x=[-3, 3, 3, -3], 
    y=[-3, -3, 3, 3], 
    z=[6, 0, -6, 0],
    i=[0, 0], j=[1, 2], k=[2, 3],
    color=color_p1, 
    opacity=0.25, 
    showscale=False, 
    hoverinfo='skip'
))

# Plane 2 (P2): -x + 2y = 0  => x = 2y (Vertical plane)
fig.add_trace(go.Mesh3d(
    x=[4, 4, -4, -4], 
    y=[2, 2, -2, -2], 
    z=[4, -4, -4, 4],
    i=[0, 0], j=[1, 2], k=[2, 3],
    color=color_p2, 
    opacity=0.25, 
    showscale=False, 
    hoverinfo='skip'
))

# 3. DRAW THE DOTTED INTERSECTION LINE
t_vals = np.array([-1.5, 1.5])
fig.add_trace(go.Scatter3d(
    x=t_vals * v[0], 
    y=t_vals * v[1], 
    z=t_vals * v[2],
    mode='lines', 
    line=dict(color='gray', width=4, dash='dot'),
    showlegend=False, 
    hoverinfo='skip'
))

# 4. ENHANCE ARROWHEADS
def add_cone(fig, start, end, color):
    fig.add_trace(go.Cone(
        x=[end[0]], y=[end[1]], z=[end[2]],
        u=[end[0]-start[0]], v=[end[1]-start[1]], w=[end[2]-start[2]],
        sizemode="absolute", sizeref=0.5, anchor="tip",  # Increased from 0.3 to 0.5 for larger arrowheads
        colorscale=[[0, color], [1, color]], showscale=False, hoverinfo='skip'
    ))

add_cone(fig, O, n1, color_p1)
add_cone(fig, O, n2, color_p2)
add_cone(fig, O, v, color_v)

# 5. DRAW RIGHT-ANGLE SYMBOL FOR P2 AND N2
w2 = np.array([2.0, 1.0, 0.0]) 
n2_hat = n2 / np.linalg.norm(n2)
w2_hat = w2 / np.linalg.norm(w2)
square_size = 0.5

# Faint guideline for w2 on the plane
fig.add_trace(go.Scatter3d(
    x=[O[0], w2[0]], y=[O[1], w2[1]], z=[O[2], w2[2]],
    mode='lines', line=dict(color='gray', width=2), showlegend=False, hoverinfo='skip'
))

# Square corner
p1_sq = O + square_size * w2_hat
p2_sq = p1_sq + square_size * n2_hat
p3_sq = O + square_size * n2_hat

fig.add_trace(go.Scatter3d(
    x=[p1_sq[0], p2_sq[0], p3_sq[0]], y=[p1_sq[1], p2_sq[1], p3_sq[1]], z=[p1_sq[2], p2_sq[2], p3_sq[2]],
    mode='lines', line=dict(color='black', width=2), showlegend=False, hoverinfo='skip'
))

# 6. MANUAL LABEL POSITIONING
label_x = [n1[0]+0.1, n2[0]-0.3, v[0]-0.2, -3.2, 4.2]
label_y = [n1[1]-0.3, n2[1]+0.3, v[1]-0.2, 3.2, 2.2]
label_z = [n1[2]+0.4, n2[2]+0.3, v[2]+0.4, 0.5, 4.5]

label_text = ['<i>n\u2081</i>\u20D7', '<i>n\u2082</i>\u20D7', '<i>v</i>\u20D7', '<b>P\u2081</b>', '<b>P\u2082</b>']
label_colors = [color_p1, color_p2, color_v, color_p1, color_p2]

fig.add_trace(go.Scatter3d(
    x=label_x, y=label_y, z=label_z,
    mode='text', text=label_text,
    textfont=dict(color=label_colors, size=18),
    showlegend=False, hoverinfo='skip'
))

# 7. FORMAT 3D LAYOUT
fig.update_layout(
    width=800, height=700,
    scene=dict(
        xaxis=dict(range=[-6, 6], dtick=2, title='x'),
        yaxis=dict(range=[-6, 6], dtick=2, title='y'),
        zaxis=dict(range=[-6, 6], dtick=2, title='z'),
        aspectmode='cube' 
    )
)

fig.show(scale=5)
```

We can find the directional vector $\vec v$ by taking the cross product of the normal vectors $\vec n_1$ and $\vec n_2$:
$$ \vec v = \vec n_1 \times \vec n_2 = \begin{bmatrix} 1 \\ 1 \\ 1 \end{bmatrix} \times \begin{bmatrix} -1 \\ 2 \\ 0 \end{bmatrix} = \begin{bmatrix} -2 \\ -1 \\ 3 \end{bmatrix}. $$
Hence the intersection of the two planes is the line $l$ through the origin in the direction of $\vec v$:
$$ l:\ x = -2t,\ y = -t,\ z = 3t. $$
